```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.xls")
df.head()
```

```python
df.describe()
```

```python
plt.hist(df['math score'], bins=10)
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df['math score'])
plt.title("Boxplot of Math Scores")
plt.show()
```

```python
plt.scatter(df['math score'], df['reading score'])
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.title("Math vs Reading Score")
plt.show()
```

```python
avg_scores = df[['math score','reading score','writing score']].mean()

avg_scores.plot(kind='bar')
plt.title("Average Scores")
plt.ylabel("Score")
plt.show()
```

# Mini Project 3  Data Visualization & Statistics

## Dataset Used
Students Performance Dataset

## Descriptive Statistics
The dataset includes numerical columns such as math score, reading score, and writing score.
Descriptive statistics like mean, median, standard deviation, and percentiles were calculated
to understand the overall distribution of scores.

## Visual Analysis

### Histogram
The histogram of math scores shows that most students scored between 60 and 80, indicating
average performance among the majority of students.

### Boxplot
The boxplot reveals a few lower outliers, which may represent students who performed poorly.
There are no extreme high outliers.

### Scatter Plot
The scatter plot between math and reading scores shows a positive correlation.
Students who perform well in math tend to perform well in reading.

### Bar Chart
The bar chart of average scores shows that reading and writing scores are slightly higher
than math scores.

## Conclusion
Overall, the dataset shows consistent student performance with some variation.
The presence of few outliers suggests individual performance differences,
but no major data quality issues are observed.
